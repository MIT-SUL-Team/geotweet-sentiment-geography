```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd
from datetime import date, timedelta
from script import leap_year
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
def get_date_list(year):
    base = date(year, 1, 1)
    numdays = 365 + int(leap_year(year))*1
    return [base + timedelta(days=x) for x in range(numdays)]
```

# 2012 Summary

```python
summary_df_2012 = pd.read_csv('master_2012_summary.csv')

x = get_date_list(2012)
y_geo = summary_df_2012['num_geo_posts']
y_sent = summary_df_2012['num_sent_posts']
y_common = summary_df_2012['num_common_posts']

plt.plot(x, y_geo, color='silver', label="Geotagged Posts")
plt.plot(x, y_sent, color='grey', label="Posts with Sentiment Scores")
plt.plot(x, y_common, color='green', label="Common Posts")

plt.title("Number of Posts by Day in 2012")
plt.legend(bbox_to_anchor=(1.6, 1.0), loc='upper right')

plt.show()
```

```python
avg_common_out_of_geo = (summary_df_2012['num_common_posts']/summary_df_2012['num_geo_posts']).mean(skipna=True)
avg_common_out_of_sent = (summary_df_2012['num_common_posts']/summary_df_2012['num_sent_posts']).mean(skipna=True)

pd.DataFrame(data = [[avg_common_out_of_geo, avg_common_out_of_sent]],
            columns=["common/geo", "common/sent"],
            index = [2012])
```
