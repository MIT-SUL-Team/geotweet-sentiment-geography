```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/sentiment_time_series/global_all_lang_overall_sentiment.csv')
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
```

```python
temp = df[df['lang']=='en'].reset_index(drop=True)
plt.figure(figsize=(15,7))
plt.plot(temp[temp['liwc_count_robust']>100]['date'], temp[temp['liwc_count_robust']>100]['liwc_score_robust'], label='LIWC')
plt.plot(temp[temp['emoji_count_robust']>100]['date'], temp[temp['emoji_count_robust']>100]['emoji_score_robust'], label='emoji')
plt.plot(temp[temp['hedono_count_robust']>100]['date'], 1.2*temp[temp['hedono_count_robust']>100]['hedono_score_robust'], label='Hedonometer')

plt.legend()

plt.show()
```

```python
df = pd.read_csv('../data/sentiment_time_series/global_en_geo_sentiment.csv')
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
```

```python
temp = df[df['country']=='US'].reset_index(drop=True)
temp = temp[temp['emoji_count']>1000].reset_index()
plt.figure(figsize=(15,7))

method = 'emoji'

plt.plot(temp['date'], temp['{}_score_robust_10q'.format(method)], label = '10th Percentile')
plt.plot(temp['date'], temp['{}_score_robust_50q'.format(method)], label = 'Median')
plt.plot(temp['date'], temp['{}_score_robust_90q'.format(method)], label = '90th Percentile')

plt.legend()

plt.show()
```
